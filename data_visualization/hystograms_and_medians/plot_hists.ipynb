Для работы с изображениями, найдем их гистограммы — характеристики распределения интенсивности изображения. Для этого можно воспользоваться следующим методом.

```python
def read_image(directory, img_name):
    return cv2.cvtColor(cv2.imread(f'{directory}/{img_name}'), cv2.COLOR_RGB2BGR)

def visualize(image, sz=(10, 10)):
    plt.figure(figsize=sz)
    plt.axis('off')
    plt.imshow(image)
```

```python
img = read_image('.', '0_84.7635.jpg')
```

```python
def extract_histogram(image, bins=(8, 8, 8)):
    hist = cv2.calcHist([image], [0, 1, 2], None, bins, [0, 256, 0, 256, 0, 256])
    cv2.normalize(hist, hist)
    return hist.flatten()
```

```python
visualize(img, (5, 5))
```

```python
colors = ('b','g','r')
f, axarr = plt.subplots(1, 3, figsize=(15, 4))
for i, color in enumerate(colors):
    img = read_image('.', '0_84.7635.jpg')
    hist = cv2.calcHist([img],[i], None, [256], [0,256])
    axarr[i].plot(hist, color = color)
    axarr[i].set_yticks([])
    axarr[i].scatter(hist.argmax(), 0, marker='x', s=50, label='Mode')
    axarr[i].scatter(img[:, :, i].mean(), 0, marker='o', s=50, label='Mean')
    axarr[i].scatter(np.median(img[:, :, i]), 0, marker='*', s=50, label='Median')
    axarr[i].legend()
plt.show()
```
