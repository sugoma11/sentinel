```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize(image):
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(image)
```

```python
im = cv2.imread('2022-05-30-00_00_2022-05-30-23_59_Sentinel-2_L2A_True_color.png')
visualize(im)
```

```python
visualize(im[:, :, 2])
```

```python
test_mask = np.where(im[:, :, 2]  > 30, 0, 255).astype('uint8')
cv2.imwrite('mask.jpg', test_mask)
```

```python
visualize(test_mask)
```

```python
kernel = np.ones((4, 4)).astype('uint8')
tst = cv2.erode(test_mask, kernel)
```

```python
visualize(tst)
```

```python
contours, hierarchy = cv2.findContours(tst, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
contours = sorted(contours, key=len, reverse=True)
imc = im.copy()
```

```python
visualize(cv2.drawContours(imc, [contours[0]], 0, (255, 255, 255), -1))
```

```python
sea = np.where(imc[:, :, 2] != 255, 0, 255).astype('uint8')
visualize(sea)
```

```python
seaa = cv2.bitwise_or(im, im, mask=sea)
visualize(seaa)
```

```python
visualize(seaa[:, :, 0])
```

```python
visualize(seaa[:, :, 1])
```

```python
visualize(seaa[:, :, 2])
```

```python
ships_mask = np.where(seaa[:, :, 2]  > 30, 0, 255).astype('uint8')
visualize(ships_mask)
```

```python
clouds = cv2.imread('2022-07-12-00_00_2022-07-12-23_59_Sentinel-2_L2A_True_color.png')
visualize(clouds)
```

```python
clouds_mask = np.where(clouds[:, :, 2]  < 238, 0, 255).astype('uint8')
visualize(clouds_mask)
```
